#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
# With the requests module we can get the HTML content and save into the coverpage variable
html = requests.get(url).content

# Next, we need to create a soup in order to allow BeautifulSoup to work by parsing our html code:
soup = BeautifulSoup(html, "html.parser")

# links always start with an <a href> so we need to locate this
wiki_links= soup.find_all("a")

# loop all the attributes of the link into a list
for link in wiki_links:
    if 'href' in link.attrs:
        links=link.attrs['href']
        print(links)

````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
# With the requests module we can get the HTML content and save into the coverpage variable
r1 = requests.get(url)
US_codes= r1.content

# Next, we need to create a soup in order to allow BeautifulSoup to work:
soup1 = BeautifulSoup(US_codes, 'html5lib')

# And finally, we can locate the elements we are looking for:
US_code_updated = soup1.find_all('div', class_='usctitlechanged')

# We loop through the list to get the updated tiotles in bold
titles_list = []
for i in US_code_updated:
    titles_list.append(i.get_text().strip()) # getting the text and stripping it of the \n, then append the 
                                            # individual item to a new list
print(titles_list)
````
